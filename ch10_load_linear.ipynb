```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pit
```

```python
df_summary = pd.read_csv('fall_summary_cleaned.csv', encoding='CP949')
df_summary
```

```python
pit.rc('font', family='NanumBarunGothic')

fig = pit.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)
ax.scatter(df_summary['caution'],df_summary['fall'])
ax.set_xlabel('주의')
ax.set_ylabel('낙상')

pit.title('주의와 낙상의 상관관계')
pit.show()
```

```python
X = df_summary.caution.values
Y = df_summary.fall.values

X = X.reshape(-1,1)
Y = Y.reshape(-1,1)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X,Y)

print(lr.coef_[0])

print(lr.intercept_)
```

```python

NewY = lr.coef_[0]*X +lr.intercept_

fig = pit.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)
ax.scatter(X, Y)
ax.set_xlabel('주의')
ax.set_ylabel('낙상')

pit.plot(X, NewY, color='red')
pit.title('y = {}*x + {}'.format(lr.coef_[0], lr.intercept_))
pit.show()
```

```python
import statsmodels.api as sm

res = sm.OLS(Y, sm.add_constant(X)).fit()
res.summary()
```
