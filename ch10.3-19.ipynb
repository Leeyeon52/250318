```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pit
```

```python
df_ratio = pd.read_csv('stacked_one_19.csv', dtype={'day':object, '부재':int, '미활동':int, '활동':int, '주의':int, ' 낙상':int}, encoding='CP949')
df_ratio.head()
```

```python
pit.style.use('default')
pit.rc('font', family='NanumBarunGothic')
pit.rcParams['figure.figsize'] = [10,8]
```

```python
day = df_ratio['day']
fall = df_ratio['낙상']
caution = df_ratio['주의']
action = df_ratio['활동']
non_action = df_ratio['미활동']

pit.bar(day, non_action, edgecolor='black', color='w', hatch='---', label='미활동')
pit.bar(day, action, bottom=non_action, edgecolor='black', color='0.9', hatch='...', label='활동')
pit.bar(day, caution, bottom=non_action + action, edgecolor='black', color='0.5', hatch='ooo', label='주의')
pit.bar(day, fall, bottom=non_action + action + caution, edgecolor='black', color='0', hatch='\\\\', label='낙상')

pit.legend(loc='best')
pit.xticks(rotation=45)
pit.show()
```
