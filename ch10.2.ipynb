```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pit
```

```python
df_ratio = pd.read_csv('stacked_ratio.csv',encoding='CP949')
df_ratio.head()
```

```python
pit.rc('font', family='NanumBarunGothic')
pit.rcParams['figure.figsize'] = [10, 8]
ax = df_ratio.plot.bar(stacked=True)
ax.set_xticklabels(df_ratio['deviceId'])

handles, labels =ax.get_legend_handles_labels()
ax.legend(reversed(handles),reversed(labels), loc='best')
```

```python
pit.style.use('default')
pit.rc('font', family='NanumBarunGothic')
pit.rcParams['figure.figsize']=[10,8]

deviceId = df_ratio['deviceId']
fall = df_ratio['낙상']
caution = df_ratio['주의']
action = df_ratio['활동']
non_action = df_ratio['미활동']
absent = df_ratio['부재']

pit.bar(deviceId, absent, edgecolor='black', color='w',hatch='///',label='부재')
pit.bar(deviceId, non_action, bottom=absent,  edgecolor='black', color='0.9', hatch='---', label='미활동')
pit.bar(deviceId, action, bottom=absent + non_action, edgecolor='black', color='0.8',hatch='...', label='활동')
pit.bar(deviceId, caution, bottom=absent + non_action + action, edgecolor='black', color='0.7', hatch='ooo', label='주의')
pit.bar(deviceId, fall, bottom=absent + non_action + action + caution, edgecolor='black', color='0.1', hatch='\\\\', label='낙상')

pit.legend(loc='best')
pit.show()
```

```python
pit.style.use('default')
pit.rc('font', family='NanumBarunGothic')
pit.rcParams['figure.figsize'] = [10,8]

df_not_absence = pd.read_csv('stacked_not_absence.csv',encoding='949')

deviceId = df_not_absence['deviceId']
fall = df_not_absence['낙상']
caution = df_not_absence['주의']
action = df_not_absence['활동']
non_action = df_not_absence['미활동']

pit.bar(deviceId, non_action, edgecolor='black', color='w',hatch='---',label='미활동')
pit.bar(deviceId, action, bottom=non_action, edgecolor='black', color='0.9', hatch="...", label='활동')
pit.bar(deviceId, caution, bottom=non_action + action,  edgecolor='black',color='0.8', hatch='ooo', label='주의')
pit.bar(deviceId, fall, bottom=non_action + action + caution, edgecolor='black', color='0',hatch='\\\\', label='낙상')

pit.legend(loc='best')
pit.show()
```
